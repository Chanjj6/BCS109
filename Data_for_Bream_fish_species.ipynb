```
# imports that will be used and libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df = pd.read_csv("https://raw.githubusercontent.com/Chanjj6/BCS109/master/Fish.csv")
df.head()
```

```
#checking for any nulls
df.isnull().sum()
```

```
# unique vaules
df.nunique()
```

```
df.describe()
```

```
sns.catplot(data=df , kind = 'box' , height=5, aspect=2)
```

```
sns.catplot(x='Weight',data=df,kind='count',height=5,aspect=1.1 )
```

```
#histogram
df.hist(figsize=(15,10), bins=20,xlabelsize=10, ylabelsize=10)
```

```
#scatter plot
plt.rcParams.update({'font.size': 12,'figure.figsize':(8,5)}) # set plot parameters
# pscatter plot of weight and height
sns.scatterplot(x='Weight', y='Height', data=df, color="red")
plt.title("Weight and Height",fontsize=12)
plt.show() # show plot
```

```
sns.boxplot(x="Weight", y="Length2", data=df)



```

```
corr=df.corr()
fig, ax =plt.subplots(figsize=(10,10))
sns.heatmap(data=corr,square=True, annot=True, cbar=True,ax=ax)
```
