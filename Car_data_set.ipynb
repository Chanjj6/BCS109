## Cars
In this data set you will be seeing data about all different kind of cars. What kind of car it is and the statistics of the car will be shown throughout the EDA. With this information you will be able to have all kinds of information for many different cars and also you can see the statistics on how the car is in a certain field of area.

```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/MatthewFried/Udemy/master/Day3/Day3%20Data.csv", 
names=['Symboling', 'Normalized_Losses', 'Brand', 'Fuel Type','Aspiration','Num Doors','Body Style',
'Drive Wheels','Engine Location','Wheel Base', 'Length','Width','Height','Curb Weight','Engine Type','Num Cylinders',
'Engine Size','Fuel System','Bore','Stroke','Compression_Ratio', 'Horsepower','Peak RPM','City MPG','Highway MPG','Price'])
df.head()
```

##Nulls
Checking for nulls will tell us if any information is missing in the data set.

```
df.isnull().sum()
```

##Unique values
Unique values tell us how of something only occur once throughout the data set

```
df.nunique()
```

##Statistics 
this chart will  tell us the statistics of the data set that is provided. This will tell us the average and other terms of the data set.

```
df.describe()
```

##Histogram 
A histrogram tells us how often a number is repeated during the data for a specific column of the data.

```
df.hist(figsize=(15,10), bins=20,xlabelsize=10, ylabelsize=10)
```

##Bar graph
This graph is labeled to tell us the prices of each brand. Each line represents the price of the car for that brand. There are many lines because in the data there are many cars listed for each brand which shows that there are many different prices for the car with the same price.

```
plt.bar("Price", "Brand", data = df, color = "blue")
plt.xlabel("Price")
plt.ylabel("Brand")
plt.title("Price based on brand")
plt.xticks(rotation = 90)

plt.show()
```

##Heatmap
A heatmap tells us the correlation between two different columns of the data set

```
corr=df.corr()
fig, ax =plt.subplots(figsize=(10,10))
sns.heatmap(data=corr,square=True, annot=True, cbar=True,ax=ax)
```
