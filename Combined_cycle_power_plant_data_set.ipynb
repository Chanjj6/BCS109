We were given the information of a power plant that has been recorded for 6 years (2006-2011). The graphs and charts will help explain what has happen in the power plant throughout the 6 years and if there were any problems that occurred throughout the years. This data set will be telling us about the temperature, the Exhaust vacuum, the Ambient pressure, the Relative humidty, and the Net hourly electrical energy output.

```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import missingno as msno 
import seaborn as sns 

df = pd.read_csv("https://raw.githubusercontent.com/MatthewFried/Udemy/master/Day1/Day1_Data.csv")
df.head()
```

```
df = df.rename(columns={'T':'Temperature',
                                    'V':'Exhaust_vacuum',
                                    'AP':'Ambient_pressure',
                                    'RH':'Relative_humidity',
                                    'EP':'Net_hourly_electrical_energy_output'})
```

There are no nulls in this data set which means that we have all the values for each section of the chart

```
#checking for any nulls
df.isnull().sum()
```

Unique values tell us which data only appear once throughout the entire data set.

```
#checking for any unique values
df.nunique()
```

```
df.describe()
```

##Histogram
In the histogram below it tells us the amount of times a number was frequently showing in the data provided. With these charts we can tell the most frequent numbers that show up for each section of the data. 

```
df.hist(figsize=(15,10), bins=20,xlabelsize=10, ylabelsize=10)
```

##Scatterplot
For the scatter plot below it will be telling us the relationship between the tempertature and the energy ouotput for a certain temperature. As we can see in the scatterplot as the temperature increases the electrical energy output decreases. This tells us what happens through the day at the power plant and it can show if the powerplant is functioning correctly. This plot tells us that the temperature and the net hourly electrical energy output has a negative linear relationship

```
#scatter plot
plt.rcParams.update({'font.size': 12,'figure.figsize':(8,5)}) # set plot parameters
# pscatter plot of energy out put and temperature
sns.scatterplot(x='Temperature', y='Net_hourly_electrical_energy_output', data=df, color="red")
plt.title("Temperature and energy output",fontsize=12)

plt.show() # show plot
```

##Heatmap
A heatmap tells us if the data that is being shown are correlated to each other

```
corr=df.corr()
fig, ax =plt.subplots(figsize=(10,10))
sns.heatmap(data=corr,square=True, annot=True, cbar=True,ax=ax)
```
